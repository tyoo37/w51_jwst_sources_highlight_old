```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u

ku_c_array_freq = 6.370608799601E+09*u.Hz
ku_bd_array_freq = 1.267120838083E+10*u.Hz
alma_b3_freq = 9.298234474179E+10*u.Hz

south_alma_b3_flux = 1.878246307243e-03*u.Jy
south_ku_c_array_flux = 3.019741815311e-4*u.Jy
south_ku_bd_array_flux = 2.983097627337e-4*u.Jy

north_alma_b3_flux = 5.707223521468e-4*u.Jy
north_ku_c_array_flux = 3.386743887366e-4*u.Jy
north_ku_bd_array_flux = 4.787934537883e-4*u.Jy

fig, ax = plt.subplots(figsize=(8,6))
ax.loglog([ku_c_array_freq.value, ku_bd_array_freq.value, alma_b3_freq.value], [south_ku_c_array_flux.value, south_ku_bd_array_flux.value, south_alma_b3_flux.value], marker='o', label='knot east', color='C0')
print('south_alma_b3_to_ku_bd_array', np.log10(south_alma_b3_flux.value/south_ku_bd_array_flux.value)/np.log10(alma_b3_freq.value/ku_bd_array_freq.value))
print('south_ku_bd_to_ku_c', np.log10(south_ku_bd_array_flux.value/south_ku_c_array_flux.value)/np.log10(ku_bd_array_freq.value/ku_c_array_freq.value))
print('south_alma_b3_to_ku_c', np.log10(south_alma_b3_flux.value/south_ku_c_array_flux.value)/np.log10(alma_b3_freq.value/ku_c_array_freq.value))
print('north_alma_b3_to_ku_bd', np.log10(north_alma_b3_flux.value/north_ku_bd_array_flux.value)/np.log10(alma_b3_freq.value/ku_bd_array_freq.value))
print('north_ku_bd_to_ku_c', np.log10(north_ku_bd_array_flux.value/north_ku_c_array_flux.value)/np.log10(ku_bd_array_freq.value/ku_c_array_freq.value))
print('north_alma_b3_to_ku_c', np.log10(north_alma_b3_flux.value/north_ku_c_array_flux.value)/np.log10(alma_b3_freq.value/ku_c_array_freq.value))
ax.loglog([ku_c_array_freq.value, ku_bd_array_freq.value, alma_b3_freq.value], [north_ku_c_array_flux.value, north_ku_bd_array_flux.value, north_alma_b3_flux.value], marker='o', label='knot west', color='C1')
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Flux Density (Jy)')
ax.legend()
```
